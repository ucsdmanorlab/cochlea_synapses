```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Ptruerange = np.linspace(0.01,0.9,10)
Prejectrange = np.linspace(0,1,21)
```

```python
Prejectrange = np.round(Prejectrange, 2)
```

```python
for Preject in Prejectrange:
    
```

```python
for Preject in Prejectrange:

    Pfinal = Ptruerange/(1-(1-Ptruerange)*(Preject))

    plt.plot(Ptruerange, Pfinal, 'o-')
    
plt.xlabel('Batch positive fraction')
#plt.xscale('log')
#plt.yscale('log')
plt.ylabel('Output positive fraction')
plt.legend(Prejectrange, title='P_reject')
plt.show()
```

```python


for Ptrue in Ptruerange:

    Pfinal = Ptrue/(1-(1-Ptrue)*(Prejectrange))

    plt.plot(Prejectrange, Pfinal, '*-')
plt.plot(Prejectrange, Prejectrange, color='lightgray', linewidth=4)    
plt.xlabel('Reject Probability')
#plt.xscale('log')
#plt.yscale('log')
plt.ylabel('Output positive fraction')
#leglist = [0]
leglist = np.round(Ptruerange,2)
plt.legend(leglist, title='Ptrue', bbox_to_anchor=(1.01,1))
#plt.show()
plt.xlim([0, 1])
plt.ylim([0, 1])

plt.savefig('Ptrue.pdf')  
```
